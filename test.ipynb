## 6. Determine el comportamiento a seguir con los valores faltantes. Explique si necesita remover alguna variable por la cantidad de valores faltantes que tiene. ¿Es factible eliminar todos los valores faltantes de todas las variables?

```python
# Determinar la cantidad de valores faltantes en cada variable
missing_values = data.isnull().sum()
print(missing_values)

# Determinar porcentaje de valores faltantes en cada variable
missing_percent = (data.isnull().sum() / len(data)) * 100
print(missing_percent)

# Eliminar columnas con una gran cantidad de valores faltantes (por ejemplo, más del 50%)
columns_to_drop = missing_percent[missing_percent > 50].index
data.drop(columns=columns_to_drop, inplace=True)

# Tratamiento de valores faltantes para variables numéricas
# Por ejemplo, usando la media para las variables continuas y la moda para las discretas
data[variables_continuas + variables_discretas] = data[variables_continuas + variables_discretas].fillna(data.mean())
```

## 8. Estudie si es conveniente hacer un Análisis de Componentes Principales. Recuerde que puede usar el índice KMO y el test de esfericidad de Bartlett. Haga un análisis de componentes principales con las variables numéricas, discuta los resultados e interprete los componentes.

```python
from sklearn.decomposition import PCA

# PCA para las variables numéricas
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data[variables_continuas + variables_discretas])

# Agregar los componentes principales al DataFrame
data['PCA_Component_1'] = pca_result[:, 0]
data['PCA_Component_2'] = pca_result[:, 1]

# Visualizar los componentes principales
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA_Component_1', y='PCA_Component_2', data=data)
plt.title('Análisis de Componentes Principales (PCA)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()

# Índice KMO y test de esfericidad de Bartlett
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity

kmo_all, kmo_model = calculate_kmo(data[variables_continuas + variables_discretas])
print(f'Índice KMO: {kmo_model}')
chi_square_value, p_value = calculate_bartlett_sphericity(data[variables_continuas + variables_discretas])
print(f'Chi-square value: {chi_square_value}')
print(f'p-value: {p_value}')
```
