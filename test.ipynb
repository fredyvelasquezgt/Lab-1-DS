## 2. Diga el tipo de cada una de las variables del dataset (cualitativa o categórica, cuantitativa continua, cuantitativa discreta)

```python
import pandas as pd

# Cargar los datos desde el archivo CSV
df = pd.read_csv('data.csv')

# Determinar el tipo de cada variable
variable_types = {}

for column in df.columns:
    if df[column].dtype == 'object':
        variable_types[column] = 'Cualitativa / Categórica'
    elif df[column].nunique() <= 10:
        variable_types[column] = 'Cuantitativa Discreta'
    else:
        variable_types[column] = 'Cuantitativa Continua'

# Mostrar el tipo de cada variable
for column, v_type in variable_types.items():
    print(f'{column}: {v_type}')
```

## 4. Aísle las variables numéricas de las categóricas, haga un análisis de correlación entre las mismas.

```python
import pandas as pd

# Cargar los datos desde el archivo CSV
df = pd.read_csv('data.csv')

# Determinar el tipo de cada variable
variable_types = {}

for column in df.columns:
    if df[column].dtype == 'object':
        variable_types[column] = 'Cualitativa / Categórica'
    elif df[column].nunique() <= 10:
        variable_types[column] = 'Cuantitativa Discreta'
    else:
        variable_types[column] = 'Cuantitativa Continua'

# Mostrar el tipo de cada variable
for column, v_type in variable_types.items():
    print(f'{column}: {v_type}')

# Aislar variables numéricas y categóricas
variables_numericas = [column for column, v_type in variable_types.items() if v_type.startswith('Cuantitativa')]
variables_categoricas = [column for column, v_type in variable_types.items() if v_type.startswith('Cualitativa')]

# Filtrar el DataFrame con las variables numéricas
df_numericas = df[variables_numericas]

# Análisis de correlación entre variables numéricas
correlation_matrix = df_numericas.corr()

# Visualizar la matriz de correlación como un mapa de calor
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlación de Variables Numéricas')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos desde el archivo CSV
df = pd.read_csv('data.csv')

# Determinar el tipo de cada variable
variable_types = {}

for column in df.columns:
    if df[column].dtype == 'object':
        variable_types[column] = 'Cualitativa / Categórica'
    elif df[column].nunique() <= 10:
        variable_types[column] = 'Cuantitativa Discreta'
    else:
        variable_types[column] = 'Cuantitativa Continua'

# Mostrar el tipo de cada variable
for column, v_type in variable_types.items():
    print(f'{column}: {v_type}')

# Utilizar las variables categóricas
variables_categoricas = [column for column, v_type in variable_types.items() if v_type.startswith('Cualitativa')]

# Tabla de frecuencia para cada variable categórica
for column in variables_categoricas:
    tabla_frecuencia = df[column].value_counts().reset_index()
    tabla_frecuencia.columns = [column, 'Frecuencia']
    print(f'\nTabla de Frecuencia para {column}:\n{tabla_frecuencia}')

    # Gráfico de barras para cada variable categórica
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column)
    plt.title(f'Distribución de {column}')
    plt.xlabel(column)
    plt.ylabel('Cantidad')
    plt.show()
```
