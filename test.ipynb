```python
# Distribución de la edad
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=30, kde=True)
plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de dispersión para 'Number of sexual partners' vs 'Age'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Number of sexual partners', y='Age', data=data)
plt.title('Relación entre Número de Parejas Sexuales y Edad')
plt.xlabel('Número de Parejas Sexuales')
plt.ylabel('Edad')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que ya has cargado los datos en un DataFrame llamado "data"

# Tratar los valores nulos en la columna 'Smokes'
data['Smokes'].fillna('Unknown', inplace=True)

# Convertir los valores no numéricos a valores numéricos
data['Smokes'] = data['Smokes'].map({'Yes': 1, 'No': 0, 'Unknown': -1})

# Gráfico de barras para 'Smokes'
plt.figure(figsize=(6, 4))
sns.countplot(data['Smokes'])
plt.title('Distribución de Fumadores (Smokes)')
plt.xlabel('Fumador (1: Sí, 0: No, -1: Desconocido)')
plt.ylabel('Cantidad')
plt.show()
```
